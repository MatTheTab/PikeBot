## Data Reading and Preparation

```python
df_train = read(data_file = "./Preprocessed_Data\Train\game_batch_0.npy.gz", column_names_file = "./Preprocessed_Data\Train\column_names.txt")
all_columns = df_train.columns
cols_to_leave = ["human", "elo", "color", "past_move_12", "current_move", "event", "clock", "stockfish_score_depth_8", "stockfish_difference_depth_8"]
delete_cols = list(set(all_columns) - set(cols_to_leave))

cols_change_types = ["human"]
for col in df_train.columns:
    if "stockfish" in col:
        cols_change_types.append(col)
        
new_types = ["int"]
while len(new_types) < len(cols_change_types):
    new_types.append(np.float32)

cols_to_standardize = cols_change_types[1:]
cols_to_standardize
```

```python
df = read(data_file = "./Model_Data\Train\Train_Batch_1.npy.gz", column_names_file = "./Model_Data\Train\column_names.txt")
df = depreprocess(dataset_type = "train", metadata_path = "./Model_Data/Metadata", processed_df = df, columns_to_divide = ["elo"], divide_val = [3000], cols_to_normalize = ["clock"], cols_to_standardize = cols_to_standardize, batch_number=1)
df.head()
```

```python
df_1 = read(data_file = "./Model_Data\Train\Train_Batch_1.npy.gz", column_names_file = "./Model_Data\Train\column_names.txt")
df_1 = depreprocess(dataset_type = "train", metadata_path = "./Model_Data/Metadata", processed_df = df, columns_to_divide = ["elo"], divide_val = [3000], cols_to_normalize = ["clock"], cols_to_standardize = cols_to_standardize, batch_number=1)
df_2 = read(data_file = "./Model_Data\Train\Train_Batch_2.npy.gz", column_names_file = "./Model_Data\Train\column_names.txt")
df_2 = depreprocess(dataset_type = "train", metadata_path = "./Model_Data/Metadata", processed_df = df, columns_to_divide = ["elo"], divide_val = [3000], cols_to_normalize = ["clock"], cols_to_standardize = cols_to_standardize, batch_number=1)
df_3 = read(data_file = "./Model_Data\Train\Train_Batch_3.npy.gz", column_names_file = "./Model_Data\Train\column_names.txt")
df_3 = depreprocess(dataset_type = "train", metadata_path = "./Model_Data/Metadata", processed_df = df, columns_to_divide = ["elo"], divide_val = [3000], cols_to_normalize = ["clock"], cols_to_standardize = cols_to_standardize, batch_number=1)
df_4 = read(data_file = "./Model_Data\Train\Train_Batch_4.npy.gz", column_names_file = "./Model_Data\Train\column_names.txt")
df_4 = depreprocess(dataset_type = "train", metadata_path = "./Model_Data/Metadata", processed_df = df, columns_to_divide = ["elo"], divide_val = [3000], cols_to_normalize = ["clock"], cols_to_standardize = cols_to_standardize, batch_number=1)
df_5 = read(data_file = "./Model_Data\Train\Train_Batch_5.npy.gz", column_names_file = "./Model_Data\Train\column_names.txt")
df_5 = depreprocess(dataset_type = "train", metadata_path = "./Model_Data/Metadata", processed_df = df, columns_to_divide = ["elo"], divide_val = [3000], cols_to_normalize = ["clock"], cols_to_standardize = cols_to_standardize, batch_number=1)
```
