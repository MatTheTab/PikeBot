```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_comparison(stockfish_result_path, engine_result_path):
    stockfish_lengths = np.loadtxt(stockfish_result_path, delimiter=';')
    engine_lengths = np.loadtxt(engine_result_path, delimiter=";")

    cumulative_stockfish = np.cumsum(stockfish_lengths[1])
    cumulative_engine = np.cumsum(engine_lengths[1])


    plt.plot(stockfish_lengths[0], cumulative_stockfish, label="stockfish")
    plt.plot(engine_lengths[0], cumulative_engine, label="Pikebot")

    plt.title('Cumulative Analysis of Stockfish and Pikebot Lengths')
    plt.xlabel('Length')
    plt.ylabel('Cumulative number of ended games')

    plt.legend()

    plt.show()
```

```python
stockfish_result_path = "results\MaiaChess_OptimalSimpleEngineAgent_silversuite_Fri_Dec_13_16-10-18_2024\statistics\MaiaChess_OptimalSimpleEngineAgent_game_lengths.csv"
engine_result_path = "results_final/MaiaChess_PikeBotHeuristic8_silversuite_Fri_Dec_13_10-23-03_2024/statistics/MaiaChess_PikeBotHeuristic8_game_lengths.csv"

plot_comparison(stockfish_result_path, engine_result_path)
```
